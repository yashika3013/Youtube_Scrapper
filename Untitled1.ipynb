```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd 
```

```python
# Set up the WebDriver
browser = webdriver.Chrome()
```

```python
browser.get('https://www.youtube.com/c/CodeWithHarry/videos')
```

```python
soup = BeautifulSoup(browser.page_source,'html.parser')
```

```python
data = []
elements = soup.find_all('ytd-rich-item-renderer')
```

```python
for sp in elements:
    title         = sp.find('a',class_ = 'yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    video_link    = sp.find('a',class_ = 'yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')
    duration      = sp.find('span', class_ = 'style-scope ytd-thumbnail-overlay-time-status-renderer').text.strip()
    
    data.append([title,duration,video_link])
```

```python
import pandas as pd 
df = pd.DataFrame(data, columns= ['title','duration','video_link'])
df
```

```python
df.to_csv('youtube.csv', index= False)
```
